**Data Description:**
The dataset contains information about movies and TV shows available on a netflix platform. It includes details such as show ID, type (movie or TV show), title, director, cast, country of production, date added to the platform, release year, rating, duration, listed genres, and a brief description of each title.

**Data Columns**

**Show_id:** Unique identifier for each movie or TV show.

**Type:** Identifies whether the entry is a movie or a TV show.

**Title:** Title of the movie or TV show.

**Director:** Director(s) of the movie or TV show.

**Cast:** Actors involved in the movie or TV show.

**Country:** Country where the movie or TV show was produced.

**Date_added:** Date the movie or TV show was added to the streaming platform.

**Release_year:** Actual release year of the movie or TV show.

**Rating:** TV rating assigned to the movie or TV show.

**Duration:** Total duration of the movie (in minutes) or number of seasons for TV shows.

**Listed_in:** Genres/categories assigned to the movie or TV show.

**Description:** Brief summary description of the movie or TV show.

**Analysis Scope:**
The dataset provides valuable insights into the platform's content library, including the types of content available, their genres, ratings, and other relevant details. The analysis aims to uncover trends, preferences, and patterns within the dataset to inform strategic decision-making for content acquisition, user engagement, and platform growth

```python
#Analysis of actors/directors of different types of shows/movies.

# Group the data by 'Type' (either 'Movie' or 'TV Show')
new_type = df.groupby('type')

# Analyze actors/directors for each group
for i, group in new_type:
    print(f"\nType: {i}\n")
    
    # Analyze actors
    print("Top 5 Actors:")
    top_actors = group['cast'].str.split(', ', expand=True).stack().value_counts().head(5)
    print(top_actors)
    
    # Analyze directors
    print("\nTop 5 Directors:")
    top_directors = group['director'].value_counts().head(5)
    print(top_directors)
```

```python
#Does Netflix has more focus on TV Shows than movies in recent years
df['date_added'] = pd.to_datetime(df['date_added'])
df['release_year'] = df['date_added'].dt.year
content_by_year = df.groupby(['type', 'release_year']).size()

# Plot the number of movies and TV shows released each year
plt.figure(figsize=(10, 6))
content_by_year.plot(kind='bar', stacked=True)
plt.title('Number of Movies and TV Shows Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig(fname="Number of Movies and TV Shows Added to Netflix Over Time.png")
plt.show()
```

**Observations:By observing the bar plot netflix has more focus on movies than TV shows.**

```python
#Which actors/actresses appear most frequently in Netflix content?

cast_split = df['cast'].str.split(', ', expand=True)
all_actors = cast_split.stack()
actor_counts = all_actors.value_counts()
top_10_actors = actor_counts.head(10)
print("Top 10 Actors/Actresses with the Most Appearances:\n")
print(top_10_actors)
```

```python
#What are the top 10 directors with the highest number of titles in the dataset?
highest_director = df[df['director'] != 'Unknown director']['director'].value_counts()
top = highest_director.idxmax()
top_director_count = highest_director.max()
plt.figure(figsize=(10, 6))
highest_director.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Directors with the Most Titles')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.savefig("TOP10directors_titles.png")
plt.show()
```

```python
#What is the average duration of movies versus TV shows on Netflix?
df['duration'] = df['duration'].astype(str)
df['duration_numeric'] = df['duration'].str.extract('(\d+)').astype(float)

average_duration = df.groupby('type')['duration_numeric'].mean()

print("Average Duration of Movies versus TV Shows:\n")
print(average_duration)
```

**Obsetvations: The Average durations of movie is approximately 99.58 minutes.This tells that the average movies tend to be 99 to 100 minutes in length.
The Average durations of TV shows is aproximately 1.76 seasons.This tells that the average TV shows to be 1 to 2 seasons.**

```python
#What are the most common genres of movies and TV shows on Netflix? 
genres_split = df['listed_in'].str.split(', ', expand=True)
all_genres = genres_split.stack()
genre_counts = all_genres.value_counts()
print("Top 10 Most Common Genres:\n")
print(genre_counts.head(10))
```
